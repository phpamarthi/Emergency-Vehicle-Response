```python
col1 = ['index',
'Make',
'Model',
'Model Year',
'Mileage',
'ADS Equipped?',
'Incident Date',
'Incident Time (24:00)',
'Roadway Type',
'Roadway Surface',
'Posted Speed Limit (MPH)',
'Lighting',
'Crash With',
'Highest Injury Severity',
'SV Pre-Crash Movement',
'SV Precrash Speed (MPH)'
]
auto = pd.DataFrame(auto, columns = col1)
auto
```

```python
auto.shape
```

```python
# compute mean and std of `Posted speed limit`
speed_mean = auto['Posted Speed Limit (MPH)'].mean()
speed_std = auto['Posted Speed Limit (MPH)'].std()

# number of NaN in `Posted speed limit` column
num_na = auto['Posted Speed Limit (MPH)'].isna().sum()

# generate `num_na` samples from N(speed_mean, speed_std**2) distribution
rand_vals = speed_mean + speed_std * np.random.randn(num_na)

# replace missing values with `rand_vals`
auto.loc[auto['Posted Speed Limit (MPH)'].isna(), 'Posted Speed Limit (MPH)'] = rand_vals
```

```python
auto['Highest Injury Severity'].value_counts()
```

```python
#Echo or delta = fatality
#Charlie = Serious 
#Bravo = Moderate
#Alpha = minor
##Omega = No Injuries Reported
```

```python
auto['Highest Injury Severity'] = auto['Highest Injury Severity'].str.replace("No Injuries Reported", "Omega")
auto['Highest Injury Severity'] = auto['Highest Injury Severity'].str.replace("Minor", "Alpha")
auto['Highest Injury Severity'] = auto['Highest Injury Severity'].str.replace("Moderate", "Bravo")
auto['Highest Injury Severity'] = auto['Highest Injury Severity'].str.replace("Fatality", "Echo/Delta")
auto['Highest Injury Severity'] = auto['Highest Injury Severity'].str.replace("Serious", "Charlie")
auto['Highest Injury Severity'].value_counts()
```
